```python
data="C:/Users/asheth13/Documents/Python_Scripts/AOC/Data_AOC_12_p1.txt" # import data
```

```python
a=[]
with open(data,"r") as file1: 
        for line in file1:
                line=tuple(line.strip("\n").split("-"))
                a.append(line)
```

```python
starts=[]
rest=[]
paths=[]

for i in range(len(a)):
    if a[i][0]=="start":
        starts.append(a[i])
        paths.append(a[i])
    elif a[i][1]=="start":
        starts.append((a[i][1],a[i][0]))
        paths.append((a[i][1],a[i][0]))
    elif a[i][1]=="end":
        rest.append(a[i])
    elif a[i][0]=="end":
        rest.append((a[i][1],a[i][0]))
    else:
        rest.append(a[i])
        rest.append((a[i][1],a[i][0]))
```

```python
small_caves=set()
for i in range(len(rest)):
    if rest[i][0]==rest[i][0].lower():
        small_caves.add(rest[i][0])
```

```python
paths_end=[]
for k in range(2*len(small_caves)+2):  
    print(range(len(paths)))
    for i in range(len(paths)):     
        for j in range(len(rest)):
            if paths[i][-1]==rest[j][0]:
                proxy=list(paths[i])
                proxy.append(rest[j][1])
                proxy=tuple(proxy)
                k=1
                for x in small_caves:
                    if proxy.count(x) <= 1:
                        k=k*1
                    else:
                        k=k*0
                if  k==1 and  proxy[-1]=="end" and (proxy not in paths_end):
                        paths_end.append(proxy)
                elif k==1 and proxy[-1]!="end" and (proxy not in paths):   
                        paths.append(proxy)
                        
```
